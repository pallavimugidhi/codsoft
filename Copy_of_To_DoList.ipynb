```python
# Google Colab To-Do List Application

tasks = []

# Load tasks from file (if exists)
def load_tasks():
    global tasks
    try:
        with open("tasks.txt", "r") as file:
            for line in file:
                text, status = line.strip().split("|")
                tasks.append({"text": text, "completed": status == "True"})
    except FileNotFoundError:
        pass

# Save tasks to file
def save_tasks():
    with open("tasks.txt", "w") as file:
        for task in tasks:
            file.write(task["text"] + "|" + str(task["completed"]) + "\n")

def show_tasks():
    if not tasks:
        print("\nNo tasks found.\n")
    else:
        print("\nYour To-Do List:")
        for i, task in enumerate(tasks, start=1):
            status = "✔" if task["completed"] else "✘"
            print(f"{i}. {task['text']} [{status}]")
        print()

def add_task():
    task = input("Enter task: ")
    tasks.append({"text": task, "completed": False})
    save_tasks()
    print("Task added successfully!\n")

def delete_task():
    show_tasks()
    try:
        num = int(input("Enter task number to delete: "))
        tasks.pop(num - 1)
        save_tasks()
        print("Task deleted!\n")
    except:
        print("Invalid selection!\n")

def mark_completed():
    show_tasks()
    try:
        num = int(input("Enter task number to mark completed: "))
        tasks[num - 1]["completed"] = True
        save_tasks()
        print("Task marked as completed!\n")
    except:
        print("Invalid selection!\n")

def update_task():
    show_tasks()
    try:
        num = int(input("Enter task number to update: "))
        new_text = input("Enter new task text: ")
        tasks[num - 1]["text"] = new_text
        save_tasks()
        print("Task updated!\n")
    except:
        print("Invalid selection!\n")

# Load previous tasks
load_tasks()

# Main Program Loop
while True:
    print("===== TO-DO LIST MENU =====")
    print("1. Show Tasks")
    print("2. Add Task")
    print("3. Update Task")
    print("4. Mark as Completed")
    print("5. Delete Task")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        show_tasks()
    elif choice == "2":
        add_task()
    elif choice == "3":
        update_task()
    elif choice == "4":
        mark_completed()
    elif choice == "5":
        delete_task()
    elif choice == "6":
        print("Goodbye!")
        break
    else:
        print("Invalid choice!\n")
```
